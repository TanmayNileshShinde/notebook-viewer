```python
## Q4. Develop an Android application to implement ProgressBar and show loading status.

## MainActivity.java

package com.example.tsprq4;

import android.os.Bundle;
import android.os.Handler;
import android.widget.Button;
import android.widget.ProgressBar;
import android.widget.TextView;

import androidx.appcompat.app.AppCompatActivity;

public class MainActivity extends AppCompatActivity {
    int progress = 0;
    @Override
    protected void onCreate(Bundle b) {
        super.onCreate(b);
        setContentView(R.layout.activity_main);

        ProgressBar pb = findViewById(R.id.progressBar);
        TextView  tv = findViewById(R.id.txt);
        Button btn = findViewById(R.id.btn);

        Handler handler = new Handler();

        btn.setOnClickListener(v -> {
            progress = 0;

            handler.postDelayed(new Runnable() {
                @Override
                public void run() {
                    if (progress <= 100) {
                        pb.setProgress(progress);
                        tv.setText("Loading " + progress + "%");

                        progress += 10;
                        handler.postDelayed(this, 500);
                    }
                }
            }, 500);
        });
    }
}

## activity_main.xml

<LinearLayout xmlns:android="http://schemas.android.com/apk/res/android"
    android:layout_width="match_parent"
    android:layout_height="match_parent"
    android:orientation="vertical"
    android:gravity="center"
    android:padding="20dp">

    <ProgressBar
        android:id="@+id/progressBar"
        android:layout_width="match_parent"
        android:layout_height="wrap_content"
        style="?android:attr/progressBarStyleHorizontal"
        android:max="100"/>

    <Button
        android:id="@+id/btn"
        android:layout_width="match_parent"
        android:layout_height="wrap_content"
        android:text="Start Loading"
        android:layout_marginTop="20dp"/>

    <TextView
        android:id="@+id/txt"
        android:layout_width="wrap_content"
        android:layout_height="wrap_content"
        android:text="Loading 0%"
        android:textSize="18sp"/>

</LinearLayout>
```

```python
## Q6. Develop an Android application using Linear Layout and Constraint Layout to create a Registration Form with Custom Toast message.

## MainActivity.java

package com.example.tsprq6;

import android.os.Bundle;
import android.view.View;
import android.widget.Button;
import android.widget.EditText;
import android.widget.TextView;
import android.widget.Toast;

import androidx.appcompat.app.AppCompatActivity;

public class MainActivity extends AppCompatActivity {

    @Override
    protected void onCreate(Bundle b) {
        super.onCreate(b);
        setContentView(R.layout.activity_main);

        EditText name = findViewById(R.id.name);
        EditText email = findViewById(R.id.email);
        EditText pass = findViewById(R.id.pass);
        Button btn = findViewById(R.id.register);

        btn.setOnClickListener(v -> {

            if (name.getText().toString().isEmpty() ||
                    email.getText().toString().isEmpty() ||
                    pass.getText().toString().isEmpty()) {

                Toast.makeText(this, "Fill all fields", Toast.LENGTH_SHORT).show();
            }
            else {
                showCustomToast("Registration Successful");
            }
        });
    }

    void showCustomToast(String msg) {
        View v = getLayoutInflater().inflate(R.layout.toast, null);
        ((TextView) v.findViewById(R.id.txt)).setText(msg);

        Toast t = new Toast(this);
        t.setView(v);
        t.setDuration(Toast.LENGTH_SHORT);
        t.show();
    }
}

## activity_main.xml

<LinearLayout xmlns:android="http://schemas.android.com/apk/res/android"
    xmlns:app="http://schemas.android.com/apk/res-auto"
    android:orientation="vertical"
    android:padding="20dp"
    android:gravity="center"
    android:layout_width="match_parent"
    android:layout_height="match_parent">

    <androidx.constraintlayout.widget.ConstraintLayout
        android:layout_width="match_parent"
        android:layout_height="wrap_content">

        <EditText
            android:id="@+id/name"
            android:hint="Name"
            android:layout_width="0dp"
            android:layout_height="wrap_content"
            app:layout_constraintTop_toTopOf="parent"
            app:layout_constraintStart_toStartOf="parent"
            app:layout_constraintEnd_toEndOf="parent"/>

        <EditText
            android:id="@+id/email"
            android:hint="Email"
            android:layout_width="0dp"
            android:layout_height="wrap_content"
            app:layout_constraintTop_toBottomOf="@id/name"
            app:layout_constraintStart_toStartOf="parent"
            app:layout_constraintEnd_toEndOf="parent"/>

        <EditText
            android:id="@+id/pass"
            android:hint="Password"
            android:inputType="textPassword"
            android:layout_width="0dp"
            android:layout_height="wrap_content"
            app:layout_constraintTop_toBottomOf="@id/email"
            app:layout_constraintStart_toStartOf="parent"
            app:layout_constraintEnd_toEndOf="parent"/>

        <Button
            android:id="@+id/register"
            android:text="Register"
            android:layout_width="0dp"
            android:layout_height="wrap_content"
            app:layout_constraintTop_toBottomOf="@id/pass"
            app:layout_constraintStart_toStartOf="parent"
            app:layout_constraintEnd_toEndOf="parent"/>

    </androidx.constraintlayout.widget.ConstraintLayout>

</LinearLayout>

## toast.xml

<LinearLayout xmlns:android="http://schemas.android.com/apk/res/android"
    android:orientation="horizontal"
    android:padding="12dp"
    android:gravity="center_vertical"
    android:background="@drawable/toast_bg"
    android:layout_width="wrap_content"
    android:layout_height="wrap_content">

    <ImageView
        android:id="@+id/icon"
        android:src="@drawable/tick"
        android:layout_width="24dp"
        android:layout_height="24dp"
        android:layout_marginEnd="8dp"/>

    <TextView
        android:id="@+id/txt"
        android:text="Success"
        android:textColor="#FFFFFF"
        android:textSize="16sp"
        android:textStyle="bold"
        android:layout_width="wrap_content"
        android:layout_height="wrap_content"/>

</LinearLayout>

## toast_bg.xml

<shape xmlns:android="http://schemas.android.com/apk/res/android">
    <solid android:color="#4CAF50"/>
    <corners android:radius="12dp"/>
</shape>
```

```python
## Q9. Develop a Calculator application using Grid Layout and basic arithmetic operations.

## MainActivity.java

package com.example.tsprq9;

import android.os.Bundle;
import android.widget.EditText;
import androidx.appcompat.app.AppCompatActivity;

public class MainActivity extends AppCompatActivity {

    EditText et;
    String n1 = "", n2 = "", op = "";
    boolean second = false;

    @Override
    protected void onCreate(Bundle b) {
        super.onCreate(b);
        setContentView(R.layout.activity_main);

        et = findViewById(R.id.et);

        findViewById(R.id.b0).setOnClickListener(v -> addNum("0"));
        findViewById(R.id.b1).setOnClickListener(v -> addNum("1"));
        findViewById(R.id.b2).setOnClickListener(v -> addNum("2"));
        findViewById(R.id.b3).setOnClickListener(v -> addNum("3"));
        findViewById(R.id.b4).setOnClickListener(v -> addNum("4"));
        findViewById(R.id.b5).setOnClickListener(v -> addNum("5"));
        findViewById(R.id.b6).setOnClickListener(v -> addNum("6"));
        findViewById(R.id.b7).setOnClickListener(v -> addNum("7"));
        findViewById(R.id.b8).setOnClickListener(v -> addNum("8"));
        findViewById(R.id.b9).setOnClickListener(v -> addNum("9"));

        findViewById(R.id.badd).setOnClickListener(v -> setOp("+"));
        findViewById(R.id.bsub).setOnClickListener(v -> setOp("-"));
        findViewById(R.id.bmul).setOnClickListener(v -> setOp("*"));
        findViewById(R.id.bdiv).setOnClickListener(v -> setOp("/"));

        findViewById(R.id.bc).setOnClickListener(v -> {
            n1 = "";
            n2 = "";
            op = "";
            second = false;
            et.setText("");
        });

        findViewById(R.id.beq).setOnClickListener(v -> {
            if (!n1.isEmpty() && !n2.isEmpty()) {
                double x = Double.parseDouble(n1);
                double y = Double.parseDouble(n2);
                double r = 0;

                if (op.equals("+")) r = x + y;
                if (op.equals("-")) r = x - y;
                if (op.equals("*")) r = x * y;
                if (op.equals("/")) r = x / y;

                et.setText(String.valueOf(r));
            }
        });
    }

    void addNum(String n) {
        if (!second) {
            n1 += n;
            et.setText(n1);
        } else {
            n2 += n;
            et.setText(n1 + op + n2);
        }
    }

    void setOp(String o) {
        if (!n1.isEmpty()) {
            op = o;
            second = true;
            et.setText(n1 + op);
        }
    }
}

## activity_main.xml

<LinearLayout xmlns:android="http://schemas.android.com/apk/res/android"
    android:orientation="vertical"
    android:padding="16dp"
    android:gravity="center"
    android:layout_width="match_parent"
    android:layout_height="match_parent">

    <EditText
        android:id="@+id/et"
        android:hint="Enter Expression"
        android:textSize="24sp"
        android:layout_width="match_parent"
        android:layout_height="wrap_content"/>

    <GridLayout
        android:layout_width="match_parent"
        android:layout_height="wrap_content"
        android:columnCount="4"
        android:rowCount="4">

        <Button
            android:id="@+id/b1"
            android:text="1" />
        <Button
            android:id="@+id/b2"
            android:text="2" />
        <Button
            android:id="@+id/b3"
            android:text="3" />
        <Button
            android:id="@+id/badd"
            android:text="+" />

        <Button
            android:id="@+id/b4"
            android:text="4" />
        <Button
            android:id="@+id/b5"
            android:text="5" />
        <Button
            android:id="@+id/b6"
            android:text="6" />
        <Button
            android:id="@+id/bsub"
            android:text="-" />

        <Button
            android:id="@+id/b7"
            android:text="7" />
        <Button
            android:id="@+id/b8"
            android:text="8" />
        <Button
            android:id="@+id/b9"
            android:text="9" />
        <Button
            android:id="@+id/bmul"
            android:text="*" />

        <Button
            android:id="@+id/bc"
            android:text="C" />
        <Button
            android:id="@+id/b0"
            android:text="0" />
        <Button
            android:id="@+id/beq"
            android:text="=" />
        <Button
            android:id="@+id/bdiv"
            android:text="/" />

    </GridLayout>

</LinearLayout>
```

```python
## Q18. Develop an Android application using Broadcast Receiver to switch modes (Airplane/Silent/Loud).

## MainActivity.java

package com.example.tsprq18;

import android.content.Intent;
import android.content.IntentFilter;
import android.media.AudioManager;
import android.os.Bundle;
import android.provider.Settings;
import android.widget.TextView;

import androidx.appcompat.app.AppCompatActivity;

public class MainActivity extends AppCompatActivity {

    ModeReceiver r;

    @Override
    protected void onCreate(Bundle b) {
        super.onCreate(b);
        setContentView(R.layout.activity_main);

        TextView tv = findViewById(R.id.txtStatus);
        AudioManager a = (AudioManager) getSystemService(AUDIO_SERVICE);

        findViewById(R.id.btnSilent).setOnClickListener(v -> {
            a.setRingerMode(AudioManager.RINGER_MODE_SILENT);
            tv.setText("Silent Mode");
        });

        findViewById(R.id.btnVibrate).setOnClickListener(v -> {
            a.setRingerMode(AudioManager.RINGER_MODE_VIBRATE);
            tv.setText("Vibrate Mode");
        });

        findViewById(R.id.btnLoud).setOnClickListener(v -> {
            a.setRingerMode(AudioManager.RINGER_MODE_NORMAL);
            tv.setText("Loud Mode");
        });

        findViewById(R.id.btnAirplaneSettings).setOnClickListener(v ->
                startActivity(new Intent(Settings.ACTION_AIRPLANE_MODE_SETTINGS)));

        r = new ModeReceiver(on -> tv.setText("Airplane Mode: " + (on ? "ON" : "OFF")));

        tv.setText("Ready");
    }

    @Override
    protected void onStart() {
        super.onStart();
        registerReceiver(r, new IntentFilter(Intent.ACTION_AIRPLANE_MODE_CHANGED));
    }

    @Override
    protected void onStop() {
        super.onStop();
        unregisterReceiver(r);
    }
}

## ModeReceiver.java

package com.example.tsprq18;

import android.content.BroadcastReceiver;
import android.content.Context;
import android.content.Intent;

public class ModeReceiver extends BroadcastReceiver {

    interface AirplaneListener {
        void onAirplaneChanged(boolean on);
    }
    AirplaneListener l;
    ModeReceiver(AirplaneListener l) {
        this.l = l;
    }

    @Override
    public void onReceive(Context c, Intent i) {
        if (l != null)
            l.onAirplaneChanged(i.getBooleanExtra("state", false));
    }
}

## activity_main.xml

<?xml version="1.0" encoding="utf-8"?>
<LinearLayout xmlns:android="http://schemas.android.com/apk/res/android"
    android:orientation="vertical"
    android:padding="16dp"
    android:gravity="center"
    android:layout_width="match_parent"
    android:layout_height="match_parent">

    <TextView
        android:id="@+id/txtStatus"
        android:text="Status will show here"
        android:textSize="18sp"
        android:layout_width="match_parent"
        android:layout_height="wrap_content"/>

    <Button
        android:id="@+id/btnSilent"
        android:text="Silent Mode"
        android:layout_width="match_parent"
        android:layout_height="wrap_content"/>

    <Button
        android:id="@+id/btnVibrate"
        android:text="Vibrate Mode"
        android:layout_width="match_parent"
        android:layout_height="wrap_content"/>

    <Button
        android:id="@+id/btnLoud"
        android:text="Loud Mode"
        android:layout_width="match_parent"
        android:layout_height="wrap_content"/>

    <Button
        android:id="@+id/btnAirplaneSettings"
        android:text="Open Airplane Mode Settings"
        android:layout_width="match_parent"
        android:layout_height="wrap_content"/>
</LinearLayout>

## AndroidManifest.xml

    <uses-permission android:name="android.permission.MODIFY_AUDIO_SETTINGS"/>
    <uses-permission android:name="android.permission.ACCESS_NOTIFICATION_POLICY"/>
```

```python
## Q19. Develop a Registration Application to insert and retrieve data using SQLite/Firebase database.

## MainActivity.java

package com.example.tsprq19;

import android.os.Bundle;
import android.widget.EditText;
import android.widget.TextView;
import android.widget.Toast;

import androidx.appcompat.app.AppCompatActivity;

public class MainActivity extends AppCompatActivity {

    @Override
    protected void onCreate(Bundle b) {
        super.onCreate(b);
        setContentView(R.layout.activity_main);

        EditText name = findViewById(R.id.name);
        EditText email = findViewById(R.id.email);
        TextView tv = findViewById(R.id.tv);

        DBHelper db = new DBHelper(this);

        findViewById(R.id.save).setOnClickListener(v -> {
            db.insert(name.getText().toString(), email.getText().toString());
            Toast.makeText(this, "Saved", Toast.LENGTH_SHORT).show();
        });

        findViewById(R.id.view).setOnClickListener(v -> {
            tv.setText(db.getData());
        });
    }
}

## DBHelper.java

package com.example.tsprq19;

import android.content.*;
import android.database.Cursor;
import android.database.sqlite.*;

public class DBHelper extends SQLiteOpenHelper {

    public DBHelper(Context c) {
        super(c, "MyDB", null, 1);
    }

    public void onCreate(SQLiteDatabase db) {
        db.execSQL("CREATE TABLE users(name TEXT, email TEXT)");
    }

    public void onUpgrade(SQLiteDatabase db, int o, int n) {}

    public void insert(String name, String email) {
        SQLiteDatabase db = this.getWritableDatabase();
        db.execSQL("INSERT INTO users VALUES(?,?)", new Object[]{name, email});
    }

    public String getData() {
        Cursor c = getReadableDatabase().rawQuery("SELECT * FROM users", null);
        String data = "";
        while (c.moveToNext())
            data += c.getString(0) + " - " + c.getString(1) + "\n";
        return data;
    }
}

## activity_main.xml

<LinearLayout xmlns:android="http://schemas.android.com/apk/res/android"
    android:layout_width="match_parent"
    android:layout_height="match_parent"
    android:orientation="vertical"
    android:padding="20dp"
    android:gravity="center">

    <EditText
        android:id="@+id/name"
        android:hint="Name"
        android:layout_width="match_parent"
        android:layout_height="wrap_content"/>

    <EditText
        android:id="@+id/email"
        android:hint="Email"
        android:layout_width="match_parent"
        android:layout_height="wrap_content"/>

    <Button
        android:id="@+id/save"
        android:text="Save"
        android:layout_width="match_parent"
        android:layout_height="wrap_content"/>

    <Button
        android:id="@+id/view"
        android:text="View"
        android:layout_width="match_parent"
        android:layout_height="wrap_content"/>

    <TextView
        android:id="@+id/tv"
        android:text="Data"
        android:layout_width="wrap_content"
        android:layout_height="wrap_content"/>

</LinearLayout>
```

```python
## Q20. Develop an Authentication Application using database concepts.

## MainActivity.java

package com.example.tsprq20;

import android.os.Bundle;
import android.widget.EditText;
import android.widget.Toast;
import androidx.appcompat.app.AppCompatActivity;

public class MainActivity extends AppCompatActivity {

    @Override
    protected void onCreate(Bundle b) {
        super.onCreate(b);
        setContentView(R.layout.activity_main);

        EditText user = findViewById(R.id.user);
        EditText pass = findViewById(R.id.pass);

        DBHelper db = new DBHelper(this);

        findViewById(R.id.login).setOnClickListener(v -> {

            String u = user.getText().toString();
            String p = pass.getText().toString();

            if (db.checkUser(u, p))
                Toast.makeText(this, "Login Success", Toast.LENGTH_SHORT).show();
            else
                Toast.makeText(this, "Invalid User", Toast.LENGTH_SHORT).show();
        });
    }
}

## DBHelper.java

package com.example.tsprq20;

import android.content.*;
import android.database.Cursor;
import android.database.sqlite.*;

public class DBHelper extends SQLiteOpenHelper {

    public DBHelper(Context c) {
        super(c, "AuthDB", null, 1);
    }

    public void onCreate(SQLiteDatabase db) {
        db.execSQL("CREATE TABLE users(username TEXT, password TEXT)");
        db.execSQL("INSERT INTO users VALUES('admin','1234')");
    }

    public void onUpgrade(SQLiteDatabase db, int o, int n) {}

    public boolean checkUser(String u, String p) {
        Cursor c = getReadableDatabase().rawQuery(
                "SELECT * FROM users WHERE username=? AND password=?",
                new String[]{u, p}
        );
        return c.moveToFirst();
    }
}

## activity_main.xml

<LinearLayout xmlns:android="http://schemas.android.com/apk/res/android"
    android:layout_width="match_parent"
    android:layout_height="match_parent"
    android:orientation="vertical"
    android:padding="20dp"
    android:gravity="center">

    <EditText
        android:id="@+id/user"
        android:hint="Username"
        android:layout_width="match_parent"
        android:layout_height="wrap_content"/>

    <EditText
        android:id="@+id/pass"
        android:hint="Password"
        android:inputType="textPassword"
        android:layout_width="match_parent"
        android:layout_height="wrap_content"/>

    <Button
        android:id="@+id/login"
        android:text="Login"
        android:layout_width="match_parent"
        android:layout_height="wrap_content"/>

</LinearLayout>
```

```python
## Q21. Develop a MyContacts application using database operations.

## MainActivity.java

package com.example.tsprq21;

import android.os.Bundle;
import android.widget.EditText;
import android.widget.TextView;
import android.widget.Toast;

import androidx.appcompat.app.AppCompatActivity;

public class MainActivity extends AppCompatActivity {

    @Override
    protected void onCreate(Bundle b) {
        super.onCreate(b);
        setContentView(R.layout.activity_main);

        EditText name = findViewById(R.id.name);
        EditText phone = findViewById(R.id.phone);
        TextView tv = findViewById(R.id.tv);

        DBHelper db = new DBHelper(this);

        findViewById(R.id.add).setOnClickListener(v -> {
            db.insert(name.getText().toString(), phone.getText().toString());
            Toast.makeText(this, "Saved", Toast.LENGTH_SHORT).show();
        });

        findViewById(R.id.view).setOnClickListener(v -> {
            tv.setText(db.getData());
        });
    }
}

## DBHelper.java

package com.example.tsprq21;

import android.content.*;
import android.database.Cursor;
import android.database.sqlite.*;

public class DBHelper extends SQLiteOpenHelper {

    public DBHelper(Context c) {
        super(c, "ContactDB", null, 1);
    }

    public void onCreate(SQLiteDatabase db) {
        db.execSQL("CREATE TABLE contacts(name TEXT, phone TEXT)");
    }

    public void onUpgrade(SQLiteDatabase db, int o, int n) {}

    public void insert(String name, String phone) {
        getWritableDatabase().execSQL(
                "INSERT INTO contacts VALUES(?,?)",
                new Object[]{name, phone}
        );
    }

    public String getData() {
        Cursor c = getReadableDatabase().rawQuery("SELECT * FROM contacts", null);
        String data = "";
        while (c.moveToNext())
            data += c.getString(0) + " - " + c.getString(1) + "\n";
        return data;
    }
}

## activity_main.xml

<LinearLayout xmlns:android="http://schemas.android.com/apk/res/android"
    android:layout_width="match_parent"
    android:layout_height="match_parent"
    android:orientation="vertical"
    android:padding="20dp"
    android:gravity="center">

    <EditText
        android:id="@+id/name"
        android:hint="Name"
        android:layout_width="match_parent"
        android:layout_height="wrap_content"/>

    <EditText
        android:id="@+id/phone"
        android:hint="Phone"
        android:inputType="phone"
        android:layout_width="match_parent"
        android:layout_height="wrap_content"/>

    <Button
        android:id="@+id/add"
        android:text="Add Contact"
        android:layout_width="match_parent"
        android:layout_height="wrap_content"/>

    <Button
        android:id="@+id/view"
        android:text="View Contacts"
        android:layout_width="match_parent"
        android:layout_height="wrap_content"/>

    <TextView
        android:id="@+id/tv"
        android:text="Contacts"
        android:layout_width="wrap_content"
        android:layout_height="wrap_content"/>

</LinearLayout>
```

```python
## Q22. Develop an Android application to access Camera using permissions.

## MainActivity.java

package com.example.tsprq22;

import android.Manifest;
import android.content.Intent;
import android.content.pm.PackageManager;
import android.graphics.Bitmap;
import android.os.Bundle;
import android.provider.MediaStore;
import android.widget.ImageView;

import androidx.appcompat.app.AppCompatActivity;
import androidx.core.app.ActivityCompat;
import androidx.core.content.ContextCompat;

public class MainActivity extends AppCompatActivity {

    ImageView img;

    @Override
    protected void onCreate(Bundle b) {
        super.onCreate(b);
        setContentView(R.layout.activity_main);

        img = findViewById(R.id.img);

        findViewById(R.id.btn).setOnClickListener(v -> {

            if (ContextCompat.checkSelfPermission(this, Manifest.permission.CAMERA)
                    != PackageManager.PERMISSION_GRANTED) {

                ActivityCompat.requestPermissions(this,
                        new String[]{Manifest.permission.CAMERA}, 1);
            } else {
                openCamera();
            }
        });
    }

    void openCamera() {
        startActivityForResult(new Intent(MediaStore.ACTION_IMAGE_CAPTURE), 1);
    }

    @Override
    protected void onActivityResult(int r, int c, Intent d) {
        super.onActivityResult(r, c, d);

        if (r == 1 && c == RESULT_OK) {
            Bitmap photo = (Bitmap) d.getExtras().get("data");
            img.setImageBitmap(photo);
        }
    }
}

## activity_main.xml

<LinearLayout xmlns:android="http://schemas.android.com/apk/res/android"
    android:layout_width="match_parent"
    android:layout_height="match_parent"
    android:orientation="vertical"
    android:gravity="center"
    android:padding="20dp">

    <Button
        android:id="@+id/btn"
        android:text="Open Camera"
        android:layout_width="match_parent"
        android:layout_height="wrap_content"/>

    <ImageView
        android:id="@+id/img"
        android:layout_width="200dp"
        android:layout_height="200dp"/>

</LinearLayout>

## AndroidManifest.xml

    <uses-permission android:name="android.permission.CAMERA"/>
```

```python
## Q23. Develop an Android application to send and receive SMS.

## MainActivity.java

package com.example.tsprqfk;

import android.Manifest;
import android.content.BroadcastReceiver;
import android.content.Context;
import android.content.Intent;
import android.content.IntentFilter;
import android.os.Bundle;
import android.telephony.SmsManager;
import android.widget.Button;
import android.widget.EditText;
import android.widget.TextView;
import androidx.appcompat.app.AppCompatActivity;
import androidx.core.app.ActivityCompat;

public class MainActivity extends AppCompatActivity {

    private EditText etPhone;
    private EditText etMsg;
    private TextView tvReceived;
    private Button btnSend;

    @Override
    protected void onCreate(Bundle savedInstanceState) {
        super.onCreate(savedInstanceState);
        setContentView(R.layout.activity_main);

        etPhone = findViewById(R.id.etPhone);
        etMsg = findViewById(R.id.etMsg);
        tvReceived = findViewById(R.id.tvReceived);
        btnSend = findViewById(R.id.btnSend);

        btnSend.setOnClickListener(v -> sendSms());

        registerReceiver(smsReceiver, new IntentFilter("SMS"), Context.RECEIVER_NOT_EXPORTED);
    }

    private void sendSms() {
        String phone = etPhone.getText().toString().trim();
        String message = etMsg.getText().toString().trim();

        if (ActivityCompat.checkSelfPermission(this, Manifest.permission.SEND_SMS) != 0) {
            ActivityCompat.requestPermissions(this, new String[]{Manifest.permission.SEND_SMS}, 1);
        } else {
            SmsManager.getDefault().sendTextMessage(phone, null, message, null, null);
        }
    }

    private final BroadcastReceiver smsReceiver = new BroadcastReceiver() {
        @Override
        public void onReceive(Context context, Intent intent) {
            String from = intent.getStringExtra("from");
            String msg = intent.getStringExtra("msg");
            tvReceived.setText("From: " + from + "\nMsg: " + msg);
        }
    };
}

## SmsReceiver.java

package com.example.tsprqfk;

import android.content.BroadcastReceiver;
import android.content.Context;
import android.content.Intent;
import android.os.Build;
import android.telephony.SmsMessage;

public class SmsReceiver extends BroadcastReceiver {

    @Override
    public void onReceive(Context context, Intent intent) {
        StringBuilder messageBody = new StringBuilder();
        String sender = "";

        Object[] pdus = (Object[]) intent.getExtras().get("pdus");

        for (Object pdu : pdus) {
            SmsMessage sms;

            if (Build.VERSION.SDK_INT >= 23) {
                String format = intent.getExtras().getString("format");
                sms = SmsMessage.createFromPdu((byte[]) pdu, format);
            } else {
                sms = SmsMessage.createFromPdu((byte[]) pdu);
            }

            sender = sms.getDisplayOriginatingAddress();
            messageBody.append(sms.getMessageBody());
        }

        Intent uiIntent = new Intent("SMS");
        uiIntent.setPackage(context.getPackageName());
        uiIntent.putExtra("from", sender);
        uiIntent.putExtra("msg", messageBody.toString());
        context.sendBroadcast(uiIntent);
    }
}

## activity_main.xml

<?xml version="1.0" encoding="utf-8"?>
<LinearLayout xmlns:android="http://schemas.android.com/apk/res/android"
    android:layout_width="match_parent"
    android:layout_height="match_parent"
    android:orientation="vertical"
    android:gravity="center"
    android:padding="16dp">

    <EditText
        android:id="@+id/etPhone"
        android:layout_width="match_parent"
        android:layout_height="wrap_content"
        android:hint="Phone"
        android:inputType="phone" />

    <EditText
        android:id="@+id/etMsg"
        android:layout_width="match_parent"
        android:layout_height="wrap_content"
        android:hint="Message"
        android:minLines="3" />

    <Button
        android:id="@+id/btnSend"
        android:layout_width="match_parent"
        android:layout_height="wrap_content"
        android:text="Send" />

    <TextView
        android:id="@+id/tvReceived"
        android:layout_width="match_parent"
        android:layout_height="wrap_content"
        android:text="No messages" />

</LinearLayout>

## AndroidManifest.xml

<?xml version="1.0" encoding="utf-8"?>
<manifest xmlns:android="http://schemas.android.com/apk/res/android">

    <!-- Permissions -->
    <uses-permission android:name="android.permission.SEND_SMS"/>
    <uses-permission android:name="android.permission.RECEIVE_SMS"/>

    <application
        android:allowBackup="true"
        android:label="@string/app_name"
        android:theme="@style/Theme.Tsprqfk">

        <!-- Main Activity -->
        <activity
            android:name=".MainActivity"
            android:exported="true">
            <intent-filter>
                <action android:name="android.intent.action.MAIN"/>
                <category android:name="android.intent.category.LAUNCHER"/>
            </intent-filter>
        </activity>

        <!-- SMS Receiver -->
        <receiver
            android:name=".SmsReceiver"
            android:exported="true"
            android:permission="android.permission.BROADCAST_SMS">
            <intent-filter android:priority="1000">
                <action android:name="android.provider.Telephony.SMS_RECEIVED"/>
            </intent-filter>
        </receiver>

    </application>
</manifest>
```

```python
## Q24. Develop an Android application to send and receive Email.

## MainActivity.java

package com.example.tsprq24;

import android.content.Intent;
import android.os.Bundle;
import android.widget.EditText;

import androidx.appcompat.app.AppCompatActivity;

public class MainActivity extends AppCompatActivity {

    @Override
    protected void onCreate(Bundle b) {
        super.onCreate(b);
        setContentView(R.layout.activity_main);

        EditText to = findViewById(R.id.etTo);
        EditText sub = findViewById(R.id.etSub);
        EditText msg = findViewById(R.id.etMsg);

        findViewById(R.id.btnSendEmail).setOnClickListener(v -> {
            Intent i = new Intent(Intent.ACTION_SEND);
            i.setType("message/rfc822");
            i.putExtra(Intent.EXTRA_EMAIL, new String[]{to.getText().toString()});
            i.putExtra(Intent.EXTRA_SUBJECT, sub.getText().toString());
            i.putExtra(Intent.EXTRA_TEXT, msg.getText().toString());
            i.setPackage("com.google.android.gm");
            startActivity(Intent.createChooser(i, "Send Email"));
        });
    }
}

## activity_main.xml

<LinearLayout xmlns:android="http://schemas.android.com/apk/res/android"
    android:orientation="vertical"
    android:padding="16dp"
    android:gravity="center"
    android:layout_width="match_parent"
    android:layout_height="match_parent">

    <EditText android:id="@+id/etTo"
        android:hint="Email"
        android:inputType="textEmailAddress"
        android:layout_width="match_parent"
        android:layout_height="wrap_content"/>

    <EditText android:id="@+id/etSub"
        android:hint="Subject"
        android:layout_width="match_parent"
        android:layout_height="wrap_content"/>

    <EditText android:id="@+id/etMsg"
        android:hint="Message"
        android:minLines="3"
        android:layout_width="match_parent"
        android:layout_height="wrap_content"/>

    <Button android:id="@+id/btnSendEmail"
        android:text="Send"
        android:layout_width="match_parent"
        android:layout_height="wrap_content"/>

</LinearLayout>
```

```python
## Q27. Develop a Flashlight application using camera flash with permission handling.

## MainActivity.java

package com.example.tsprq27;

import android.Manifest;
import android.hardware.camera2.CameraManager;
import android.os.Bundle;

import androidx.appcompat.app.AppCompatActivity;
import androidx.core.app.ActivityCompat;

public class MainActivity extends AppCompatActivity {

    CameraManager cm;
    String id;

    @Override
    protected void onCreate(Bundle b) {
        super.onCreate(b);
        setContentView(R.layout.activity_main);

        cm = (CameraManager) getSystemService(CAMERA_SERVICE);

        try {
            id = cm.getCameraIdList()[0];
        } catch (Exception e) {}

        ActivityCompat.requestPermissions(this,
                new String[]{Manifest.permission.CAMERA}, 1);

        findViewById(R.id.on).setOnClickListener(v -> {
            try { cm.setTorchMode(id, true); } catch (Exception e) {}
        });

        findViewById(R.id.off).setOnClickListener(v -> {
            try { cm.setTorchMode(id, false); } catch (Exception e) {}
        });
    }
}

## activity_main.xml

<LinearLayout xmlns:android="http://schemas.android.com/apk/res/android"
    android:gravity="center"
    android:orientation="vertical"
    android:padding="20dp"
    android:layout_width="match_parent"
    android:layout_height="match_parent">

    <Button
        android:id="@+id/on"
        android:text="ON"
        android:layout_width="match_parent"
        android:layout_height="wrap_content"/>

    <Button
        android:id="@+id/off"
        android:text="OFF"
        android:layout_width="match_parent"
        android:layout_height="wrap_content"/>
</LinearLayout>

## AndroidManifest.xml

    <uses-permission android:name="android.permission.CAMERA"/>
    <uses-feature android:name="android.hardware.camera.flash"/>
```
