```python
## 14. Write a C program for factorial calculation using for loop and test it.

#include <stdio.h>

int main() {
    int n, i;
    long long fact = 1;

    printf("Enter a number: ");
    scanf("%d", &n);

    if(n < 0) {
        printf("Factorial not defined for negative numbers");
    } else {
        for(i = 1; i <= n; i++) {
            fact = fact * i;
        }
        printf("Factorial = %lld", fact);
    }

    return 0;
}

| TC ID      | Objective                           | Input  | Expected Result | Actual Result | Status |
| ---------- | ----------------------------------- | ------ | --------------- | ------------- | ------ |
| TC_FACT_01 | Verify factorial of positive number | n = 5  | 120             | 120           | Pass   |
| TC_FACT_02 | Verify factorial of 0               | n = 0  | 1               | 1             | Pass   |
| TC_FACT_03 | Verify factorial of 1               | n = 1  | 1               | 1             | Pass   |
| TC_FACT_04 | Verify negative input               | n = -3 | Error message   | Error shown   | Pass   |
| TC_FACT_05 | Verify large number                 | n = 10 | 3628800         | 3628800       | Pass   |
```
